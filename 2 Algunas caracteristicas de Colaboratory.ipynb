## Explorar la Documentación con  help() y con ``?``

El lenguaje Python y su ecosistema de data science está construido teniendo en mente al usuario/programador, y esto ser refleja también en el acceso a la documentación.
Todos los objetos Python contienen una referencia a un string, conocido como *doc string*, que proporciona un breve resumen del objeto y de cómo usarlo, en la gran mayoría de los casos .
La función predefinida de Python ``help()``  accede a esta información e imprime el resultado.

Por ejemplo, para ver la documentación de la función predefinida ``len``:


````
help(len)
````

````
# busca ayuda sobre 'range' usando help()
help(range)
````

Buscar ayuda sobre un objeto es muy común, por ello Python introduce el carácter ``?`` como abreviatura del uso de ``help()`` para acceder a su documentación que en Google Colab abre una ventana de ayuda en la parte superior derecha:


````
len?
````

````
# busca ayuda sobre 'range' usando ?
range?
````

Esta notación funciona con cualquier cosa, incluso con los métodos de los objetos:


````
L = [1, 2, 3]
L.insert?
````

e incluso con las variables en sí mismas:

````
L?
````

Funciona también con funciones u objetos creados por el usuario, si al definirlas hemos definido su *doc string*.


````
def square(a):
  """Return the square of a."""
  return a ** 2
````

Si ejecutamos square``?``:


````
square?
````

# Comandos mágicos

El Python tiene otras funcionalidades avanzadas que pueden accederse mediante los llamados *comandos mágicos*. Estos comandos se reconocen fácilmente porque vienen precedidos de un caracter ``%``.

Los comandos mágicos con prefijo ``%`` operan sobre una linea de código (*linea mágica*) y con prefijo ``%%`` sobre una celda (*celda mágica*).


##Ejecutar código externo: ``%run``

Conforme tenemos proyectos de código más grandes, podemos tener módulos en python que podemos utilizar desde Python para realizar determinados cálculos o explorar su funcionamiento interactivamente.

Esto puede hacerse con el comando mágico ``%run``.

Por ejemplo si tenemos un fichero ``myscript.py`` con el siguiente código:

```python
#-------------------------------------
# file: myscript.py

def square(x):
    """square a number"""
    return x ** 2

for N in range(1, 4):
    print(N, "squared is", square(N))
```

Podemos ejecutarlo en nuestra sesión de IPython:

```ipython
In [6]: %run myscript.py
1 squared is 1
2 squared is 4
3 squared is 9
```

Es importante notar que una vez que se ha corrido el script todas las funciones que se definían en el mismo son accesibles desde la sesion de Python:

```ipython
In [7]: square(5)
Out[7]: 25
```

## Medir el tiempo de ejecución: ``%time`` y ``%timeit``

Los comandos mágicos ``%time`` y ``%timeit`` es también muy útil, porque nos permite medir el tiempo de ejecución del código. El segundo corre el codigo varias veces para hacer una estimación más robusta del tiempo de ejecución.

Por ejemplo, para medir la creación de una lista en una *Comprehension List*:


````
 %timeit L = [n ** 2 for n in range(1000)]
````

Para medir el tiempo de ejecución de una celda se añade un segundo ``%`` al comando, esto hace que la celda sea mágica.

Por ejemplo, el mismo código implementado con un bucle ``for``:

````
%%timeit
L = []
for n in range(1000):
  L.append(n ** 2) 
````

¿Qué es más rápido?

````
# Compara las medidas de %time y %timeit sobre la lista de compresión [ x for x in range(2000) if x % 2 == 0]
%timeit [ x for x in range(2000) if x % 2 == 0]
````

#Salidas gráficas e interactivas
Hasta ahora, todos los resultados generados han sido texto, pero pueden ser más interesantes, como el cuadro a continuación.

````
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)

plt.title("Fills and Alpha Example")
plt.show()
````
